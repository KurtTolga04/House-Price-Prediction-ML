```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.model_selection import KFold, cross_val_score
from sklearn.linear_model import Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb
```

```python
# --------------------------
# 1️⃣ VERİ YÜKLEME
# --------------------------
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
# --------------------------
# 2️⃣ ORTAK TEMİZLİK FONKSİYONU
# --------------------------
# Train ve Test setine aynı işlemi uygulamak için fonksiyon tanımlıyoruz.
def clean_and_engineer(df, frontage_map=None):
    df_out = df.copy()
    
    # Gereksiz kolonları at
    drop_cols = ['PoolQC', 'MiscFeature', 'Alley', 'Fence']
    df_out.drop(columns=[c for c in drop_cols if c in df_out.columns], inplace=True)

    # Kategorik boşlukları "None" yap
    none_fill_cols = [
        'MasVnrType', 'FireplaceQu', 'GarageType', 'GarageFinish',
        'GarageQual', 'GarageCond', 'BsmtQual', 'BsmtCond',
        'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2'
    ]
    for col in none_fill_cols:
        if col in df_out.columns:
            df_out[col] = df_out[col].fillna("None")

    # Sayısal boşlukları 0 yap (Mantıklı olanlar)
    if 'MasVnrArea' in df_out.columns:
        df_out['MasVnrArea'] = df_out['MasVnrArea'].fillna(0)

    # DÜZELTME: GarageYrBlt 0 olursa model sapıtır. Garaj yoksa evin yapım yılını verelim.
    if 'GarageYrBlt' in df_out.columns:
        df_out['GarageYrBlt'] = df_out['GarageYrBlt'].fillna(df_out['YearBuilt'])

    # LotFrontage (Ön Cephe): Train setindeki mahalleye göre doldur
    # Eğer harita verilmediyse (Train aşaması), kendisi hesaplasın
    if 'LotFrontage' in df_out.columns:
        if frontage_map is None:
            # Sadece Train için çalışır
            df_out['LotFrontage'] = df_out.groupby('Neighborhood')['LotFrontage'].transform(lambda x: x.fillna(x.median()))
        else:
            # Test için: Train'den gelen değerleri kullan
            df_out['LotFrontage'] = df_out['LotFrontage'].fillna(df_out['Neighborhood'].map(frontage_map))
            # Hala boş kalan varsa genel ortalama ile doldur
            df_out['LotFrontage'] = df_out['LotFrontage'].fillna(df_out['LotFrontage'].median())

    # Elektrik sistemi (Mode ile doldur)
    if 'Electrical' in df_out.columns:
        df_out['Electrical'] = df_out['Electrical'].fillna(df_out['Electrical'].mode()[0])

    # FEATURE ENGINEERING
    # TotalSF (Toplam Alan)
    if all(col in df_out.columns for col in ['TotalBsmtSF', '1stFlrSF', '2ndFlrSF']):
        df_out['TotalBsmtSF'] = df_out['TotalBsmtSF'].fillna(0)
        df_out['1stFlrSF'] = df_out['1stFlrSF'].fillna(0)
        df_out['2ndFlrSF'] = df_out['2ndFlrSF'].fillna(0)
        df_out['TotalSF'] = df_out['TotalBsmtSF'] + df_out['1stFlrSF'] + df_out['2ndFlrSF']
    
    # HouseAge (Ev Yaşı)
    if 'YrSold' in df_out.columns and 'YearBuilt' in df_out.columns:
        df_out['HouseAge'] = df_out['YrSold'] - df_out['YearBuilt']

    return df_out
```

```python
# --------------------------
# 3️⃣ TRAIN VERİSİNİ HAZIRLA
# --------------------------
# Önce LotFrontage için Train setinden haritalama (mapping) çıkarıyoruz (Data Leakage Önleme)
frontage_map = train.groupby('Neighborhood')['LotFrontage'].median()

# Train verisini temizle
train_clean = clean_and_engineer(train, frontage_map=None) # Train olduğu için map None geçilebilir ama içeride hesaplanıyor zaten.

# Hedef Değişken Log Dönüşümü
train_clean['SalePrice_log'] = np.log1p(train_clean['SalePrice'])

# X ve y ayır
X = train_clean.drop(columns=['SalePrice', 'SalePrice_log', 'Id'])
y = train_clean['SalePrice_log']
```

```python
# --------------------------
# 4️⃣ PIPELINE & MODEL KURULUMU
# --------------------------
num_cols = X.select_dtypes(include=['int64','float64']).columns.tolist()
cat_cols = X.select_dtypes(include=['object']).columns.tolist()

# Preprocessing Pipeline
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')), # Kalan ufak tefek eksikler için
    ('scaler', StandardScaler())
])

cat_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='constant', fill_value='Missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

preprocessor = ColumnTransformer([
    ('num', num_pipeline, num_cols),
    ('cat', cat_pipeline, cat_cols)
])

# Modeller
models = {
    'Ridge': Ridge(random_state=42),
    'Lasso': Lasso(random_state=42, alpha=0.001 , max_iter=30000), # Alpha ayarı eklendi
    'RandomForest': RandomForestRegressor(random_state=42, n_jobs=-1, n_estimators=200),
    'XGBoost': xgb.XGBRegressor(random_state=42, n_jobs=-1, verbosity=0)
}
```

```python
# --------------------------
# 5️⃣ MODEL KARŞILAŞTIRMA & SEÇİM
# --------------------------
print("Modeller Cross-Validation ile Test Ediliyor...")

results = {}
best_score = float('inf')
best_model_name = ""

for name, model in models.items():
    pipe = Pipeline([('pre', preprocessor), ('model', model)])
    
    # Negatif RMSE döndüğü için eksi ile çarpıyoruz
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    cv_scores = -cross_val_score(pipe, X, y, cv=kf, scoring='neg_root_mean_squared_error')
    mean_score = cv_scores.mean()
    
    results[name] = mean_score
    print(f"{name}: RMSE = {mean_score:.4f}")
    
    if mean_score < best_score:
        best_score = mean_score
        best_model_name = name

print(f"\n🏆 EN İYİ MODEL: {best_model_name} (RMSE: {best_score:.4f})")
```

```python
# --------------------------
# 6️⃣ FİNAL MODEL EĞİTİMİ
# --------------------------
# DÜZELTME: Hardcoded model yerine, yukarıda bulduğumuz en iyi modeli kullanıyoruz.
best_model_instance = models[best_model_name]

final_pipe = Pipeline([
    ('pre', preprocessor),
    ('model', best_model_instance)
])

final_pipe.fit(X, y)
joblib.dump(final_pipe, "best_house_price_model.pkl")
print("Final model eğitildi ve kaydedildi.")
```

```python
# --------------------------
# 7️⃣ TEST & SUBMISSION
# --------------------------
# Test verisini hazırla (Aynı fonksiyonu ve Train'den öğrendiğimiz frontage_map'i kullanıyoruz)
test_clean = clean_and_engineer(test, frontage_map=frontage_map)

# ID sütununu sakla ve X_test'ten çıkar
test_ids = test_clean['Id']
X_test = test_clean.drop(columns=['Id'])

# Tahmin yap
y_pred_log = final_pipe.predict(X_test)
y_pred = np.expm1(y_pred_log) # Log'dan geriye çevir

# Dosyayı oluştur
submission = pd.DataFrame({
    "Id": test_ids,
    "SalePrice": y_pred
})
submission.to_csv("submission_corrected.csv", index=False)
print("✅ Submission dosyası hazır: submission_corrected.csv")
```

```python
# --------------------------
# 8️⃣ ÖNEMLİ ÖZELLİKLER (Feature Importance)
# --------------------------
# Eğer model ağaç tabanlıysa (Tree/Forest/XGB) önem grafiği çiz
if hasattr(best_model_instance, 'feature_importances_'):
    # Feature isimlerini al
    num_features = final_pipe.named_steps['pre'].named_transformers_['num'].get_feature_names_out()
    cat_features = final_pipe.named_steps['pre'].named_transformers_['cat'].get_feature_names_out()
    feature_names = np.concatenate([num_features, cat_features])
    
    importances = best_model_instance.feature_importances_
    feat_importance = pd.Series(importances, index=feature_names).sort_values(ascending=False).head(20)

    plt.figure(figsize=(10,6))
    feat_importance.plot(kind='barh')
    plt.gca().invert_yaxis()
    plt.title(f"Top 20 Feature Importance - {best_model_name}")
    plt.tight_layout()
    plt.show()
else:
    print(f"{best_model_name} modeli için feature importance grafiği çizilemiyor (Linear model olabilir).")
```
